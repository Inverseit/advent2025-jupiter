```python
with open("input.txt") as f:
    x = f.read()
x = x.split(",")
```

```python
def process(f):
    res = 0
    
    for r in x:
        r1 = list(map(int,r.split("-")))
        s, e = r1[0], r1[1]
        splitter = 10 ** (len(str(e)) - 1)
        if s < splitter:
            res += f(s, splitter - 1)
            res += f(splitter, e)
        else:
            res += f(s, e)
    return res
```

```python
def count2(a, b):
    digits = len(str(a))
    ans = 0
    if digits % 2 == 1:
        ans = 0
    else:
        d = 10 ** (digits // 2) + 1
        ans = sum(range(((a + d - 1) // d) * d, b + 1, d))
    return ans
    

process(count2)
```

```python
def valid_divs(n):
    return [
        (10**n - 1) // (10**(n//k) - 1)
        for k in range(2, n + 1)
        if n % k == 0
    ]

def countAll(a, b):
    n = len(str(a))  # assume a..b are same digit-length chunk
    bad = set()
    for d in valid_divs(n):
        start = ((a + d - 1) // d) * d
        bad.update(range(start, b + 1, d))
    return sum(bad)

process(countAll)
```
