```python
def process(row, k=2):
  digits = list(map(int, [c for c in row.strip() if c.isdigit()]))
  n = len(digits)

  # If there are fewer than k digits, return 0 (not possible to select k digits)
  if n < k:
    return 0

  # DP table: dp[i][j] is the maximum number we can form using the first i digits and selecting j digits
  dp = [[0] * (k + 1) for _ in range(n + 1)]

  for i in range(1, n + 1):
    for j in range(1, k + 1):
      # Option 1: Skip the current digit
      dp[i][j] = dp[i - 1][j]
      # Option 2: Include the current digit (if we have enough digits left to select)
      if j > 0:
        dp[i][j] = max(dp[i][j], dp[i - 1][j - 1] * 10 + digits[i - 1])

  # The result is the maximum number we can form using all n digits and selecting exactly k digits
  return dp[n][k]


def run(k=2):
  # Read the file and calculate the total sum
  with open('input.txt', 'r') as file:
    total_sum = sum(process(row, k) for row in file)

  print("Total Sum:", total_sum)
```

```python
run(2)
```
