```python
# Read the input file
with open('input.txt', 'r') as file:
  lines = file.readlines()

# Separate ranges and individual numbers
ranges = []
numbers = []
is_range_section = True

for line in lines:
  line = line.strip()
  if not line:  # Empty line indicates the switch between ranges and numbers
    is_range_section = False
    continue
  if is_range_section:
    ranges.append(tuple(map(int, line.split('-'))))
  else:
    numbers.append(int(line))
```

```python
ranges.sort(key=lambda x: x[0])
```

```python
count = 0
for n in numbers:
  for r in ranges:
    if r[0] <= n <= r[1]:
      count += 1
      break
print(count)
```

```python
# merge overlapping ranges using sweep line algorithm
events = []
for r in ranges:
  events.append((r[0], 'start'))
  events.append((r[1] + 1, 'end'))  # +1 to make end exclusive
events.sort()
merged_ranges = []
active_intervals = 0
current_start = None
for point, typ in events:
  if typ == 'start':
    if active_intervals == 0:
      current_start = point
    active_intervals += 1
  else:  # typ == 'end'
    active_intervals -= 1
    if active_intervals == 0:
      merged_ranges.append((current_start, point - 1))  # -1 to revert to inclusive end
      
```

```python
# Now count number of all elements in merged ranges, basically count how many integers are covered by merged ranges

total_count = 0
for r in merged_ranges:
  total_count += (r[1] - r[0] + 1)
```
